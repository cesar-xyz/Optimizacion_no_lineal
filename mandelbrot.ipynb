```python
def f(z, c, n):
    return z ** n + c
```

```python
def iterar(z0, c, iteraciones_maximas, n):
    z = z0
    for i in range(iteraciones_maximas):
        if abs(z) > 2:
            return i
        z = f(z, c, n)
    return iteraciones_maximas
```

```python
import matplotlib.pyplot as plt
import numpy as np

lado = 2
n_puntos = 200
iteraciones_maximas = 50

x = np.linspace(-lado, lado, n_puntos)
y = np.linspace(-lado, lado, n_puntos)

matriz = np.zeros((n_puntos, n_puntos))

for i in range(n_puntos):
    for j in range(n_puntos):
        z0 = complex(x[i], y[j])
        matriz[j, i] = iterar(z0, x[i] + 1j * y[j], iteraciones_maximas, 2)
```

```python
plt.imshow(matriz, cmap='twilight_r', extent=(-lado, lado, -lado, lado))
plt.xlabel('Parte real')
plt.ylabel('Parte imaginaria')
plt.show()
```
