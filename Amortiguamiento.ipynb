```python
# Dependencias a utilizar.
import control as ctrl
import matplotlib.pyplot as plt
from control.pzmap import pzmap
```

```python
s = [0, 0.2, 0.4, 0.6, 0.8, 1, 2]
wn = 1
for i in s:
    num = [pow(wn, 2)]
    den = [1, 2 * i * wn, pow(wn, 2)]
    sys = ctrl.tf(num, den)
    sys_cl = ctrl.feedback(sys, 2)

    #Respuesta al escaln
    t, y = ctrl.step_response(sys_cl)
    #Graficas
    plt.plot(t, y)
    plt.title(u"Respuesta al escaln")
    plt.xlabel("segundos")
plt.legend(s)
```

```python
for i in s:
    num = [pow(wn, 2)]
    den = [1, 2 * i * wn, pow(wn, 2)]
    sys = ctrl.tf(num, den)
    sys_cl = ctrl.feedback(sys, 2)
    plt.figure()
    polos, zeros = pzmap(sys_cl), plt.grid(True)
    plt.title(f"Pole zero map of {i}")
```
