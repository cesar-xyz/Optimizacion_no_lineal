# Funciones de transferencia: estabilidad
Determinar los ceros y los polos de las siguientes funciones de transferencia y graficarlos en el plano complejo S, decir si el sistema es Estable o Inestable.


```python
# Dependencias a utilizar.
import control as ctrl
import matplotlib.pyplot as plt
from control.pzmap import pzmap
```

## 1) $\frac{(s+1)(s-1)}{s(s+2)(s+10)}$
\
$(s+1)(s-1) = s^2-1$
\
$s(s+2)(s+10)= s^3+12s^2+20s$

```python
num = [1,0,1]
den = [1,12,20,0]
sys = ctrl.tf(num, den)
print(sys)

sys_cl = ctrl.feedback(sys, 2)

#Respuesta al escalón
t, y = ctrl.step_response(sys_cl)
```

```python
#Graficas
plt.plot(t, y)
plt.grid(True)
plt.title(u"Respuesta al escalón")
plt.xlabel("segundos")
```

```python
polos, zeros = pzmap(sys_cl), plt.grid(True)
```

#### El sistema es **estable**
